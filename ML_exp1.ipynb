```python
top_products = df.groupby("Description")["Quantity"].sum().head(10)

# Bar Chart
top_products.plot(kind='bar')
plt.title("Top 10 Products by Quantity Sold")
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.show()

# Line Chart
top_products.plot(kind='line')
plt.title("Sales Trend of Top Products")
plt.show()
```

```python
df = pd.read_csv("/content/marketing_campaign(1).csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'] = 2026 - df['Year_Birth']
```

```python
plt.figure()
df['Age'].value_counts().sort_index().plot(kind='bar')
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['Age'])
plt.title("Age Boxplot")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['Income'])
plt.title("Income Distribution")
plt.show()
```

```python
df['Income'].dropna().plot(kind='hist', bins=20)
plt.xlabel("Income")
plt.ylabel("Customers")
plt.title("Income Distribution")
plt.show()
```

```python
df['Total_Spending'] = (
    df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] +
    df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
)
```

```python
plt.figure()
sns.boxplot(x=df['Total_Spending'])
plt.title("Total Spending Distribution")
plt.show()
```
